# Introduction to Matplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
#creating axes x and y
plt.plot([1,2.5,3.7,4], [55, 65, 75,85])
plt.show()
```

```python
#Confusing way of ploting 
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
#another confusing way
x = [1,2,3,4]
y = [1, 2, 3, 4]
fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.show()
```

```python
#now doing it the easier way and we will continue to use this in the future
fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
```
