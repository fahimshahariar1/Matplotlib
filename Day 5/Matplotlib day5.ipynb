## Plotting from Pandas

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
#Removing the characters so that we can use arithemetic operations on them
car_sales["Price"] = car_sales["Price"].str.replace('[$, .]', '', regex=True)
car_sales
```

```python
#Remove the extra 0s
car_sales["Price"] = car_sales["Price"].str[:-2]
car_sales
```

```python
#We will add a date range column
car_sales["Sale Date"] = pd.date_range("1/1/2025", periods=len(car_sales))
car_sales
```

```python
#We will add another column for total sales
car_sales["Total Sales"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
#Let's plot this 
car_sales.plot(x = "Sale Date", y = "Total Sales");
```

```python
#Let's try to do a scatter plot
car_sales.plot(x = "Odometer (KM)", y = "Price", kind="scatter");
```

```python
#bar plot using pandas
car_sales.plot(x = "Make", y = "Odometer (KM)", kind = "bar");
```

```python
#then definitely we can do histogram as well
car_sales["Odometer (KM)"].plot(kind="hist");
```
