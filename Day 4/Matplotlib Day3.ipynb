## Making figures with Numpy array

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create data
x = np.linspace(0, 10, 100)
x
```

```python
#plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
#now we will use the same data to create a sactter plot
fig, ax = plt.subplots()
ax.scatter(x, x**2)
```

```python
#another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
#make a plot from dictionary
car_prices = {"Toyota": 100,
             "Honda": 200,
             "BMW": 500}

fig, ax = plt.subplots()
ax.bar(car_prices.keys(), car_prices.values())
ax.set(title="Car Prices",
      ylabel="Price");
```

```python
#creating bar in a horizontal line
fig, ax = plt.subplots()
ax.barh(list(car_prices.keys()), list(car_prices.values()))
```

```python
#Histogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

```python
#Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize = (10,5))

#Plot to each different axis

ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(car_prices.keys(), car_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
#Subplot option 2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize = (10,5))

#plot to different index

ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(car_prices.keys(), car_prices.values())
ax[1,1].hist(np.random.randn(1000))
```
