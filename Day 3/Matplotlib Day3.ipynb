## Making figures with Numpy array

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create data
x = np.linspace(0, 10, 100)
x
```

```python
#plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
#now we will use the same data to create a sactter plot
fig, ax = plt.subplots()
ax.scatter(x, x**2)
```

```python
#another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
#make a plot from dictionary
car_prices = {"Toyota": 100,
             "Honda": 200,
             "BMW": 500}

fig, ax = plt.subplots()
ax.bar(car_prices.keys(), car_prices.values())
ax.set(title="Car Prices",
      ylabel="Price");
```
