## Matplotlib Example Workflow

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
```

```python
fig, ax = plt.subplots(figsize=(10,10))

ax.plot(x,y)
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")
```
